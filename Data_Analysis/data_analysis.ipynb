```python
#import libaries 
import numpy as np
import pandas as pd
import hvplot.pandas
from pathlib import Path
```

```python
# Using the Pandas read_csv function and the Path module, reads in the data
# from the "Market_Data.csv" file and create the Pandas DataFrame
spy_prices_df = pd.read_csv(
    Path('Resources/spy_closing_prices.csv')
)
    
# def set_date_index(df,new_df):
#     columns = ['Date','new_df']
#     new_df.columns = columns
#     new_df = new_df.set_index('Date')
    
#     return(new_df)
def set_date_index(column_name, df, new_df):
    columns = ['Date',column_name]
    df.columns = columns
    new_df = df.set_index('Date')
    return(new_df)
```

```python
spy_closing_prices= set_date_index('SPY_Closing_Price', spy_prices_df, spy_closing_prices)

# set_date_index(spy_prices_df,spy_closing_prices_df)

# Review the first and last five rows of the DataFrame
display(spy_closing_prices.head())
display(spy_closing_prices.tail())
```

```python
spy_daily_returns_df = pd.read_csv(
    Path('./Resources/spy_daily_returns.csv')
)
# spy_daily_reutrns_df = set_date_index('SPY_Daily_Return', spy_daily_returns_df, spy_daily_returns_df)

columns = ['Date', 'SPY_Daily_Return']
spy_daily_returns_df.columns = columns

spy_daily_returns_df= spy_daily_returns_df.set_index('Date')
display(spy_daily_returns_df.head())
display(spy_daily_returns_df.tail())
```

```python
#crypto_prices_df = pd.read_csv(
#    Path("../Resources/Market_Data.csv")
#)

crypto_historic_closing = pd.read_csv('Resources/crypto_historic_closing.csv')
crypto_historic_daily_returns = pd.read_csv('Resources/crypto_historic_daily_returns.csv')

crypto_historic_closing = crypto_historic_closing.set_index('Date')
crypto_historic_daily_returns = crypto_historic_daily_returns.set_index('Date')

# Review the first and last five rows of the DataFrame
display(crypto_historic_closing.head())
display(crypto_historic_daily_returns.tail())
```

```python
# Create the plot for the daily returns spy
plot_daily_returns_spy = spy_daily_returns_df.hvplot.bar(
    title ='Total Daily Returns',
    y='SPY_Daily_Return',
    x='Date',
    color="blue",
    hover_color = "yellow",
    xlabel='Time',
    ylabel="Total Daily Returns",
    rot=90
)

# Visualize the plot
plot_daily_returns_spy
```

```python
plot_daily_returns_crypto = crypto_historic_daily_returns.hvplot.bar(
    x='Date',
    y=['Btc_daily_returns', 'Eth_daily_returns', 'Ltc_daily_returns'],
    height=600,
    width=1200,
    rot=90,
    stacked=True,
    title='Crypto Daily Return of BTC, ETH, LTC from 2019-7-1 - 2021-7-1'
).opts(yformatter='%.0f')

# Visualize the plot
plot_daily_returns_crypto
```

```python
# Composes an overlay plot using the '*' operator
# ax = crypto_historic_daily_returns.plot(
#     figsize=(10,7),
#     xlabel='Date',
#     ylabel="Total Daily Returns",
#     title ="Comparison of Crypto Currencies and Market Data"
#     )
# spy_daily_reutrns_df.plot(ax=ax)

# Styles the new overlay plot
overlay_plot = plot_daily_returns_crypto * plot_daily_returns_spy

overlay_plot.opts(
    height= 500,
    width=1000,
    ylabel="Total Daily Returns",
    title ="Comparison of Crypto Currencies and Market Data"
    )
```

```python
# Create new DataFrame of cummulative returns
spy_cummulative_returns = (1 + spy_daily_returns_df).cumprod()
columns = ['SPY_cum_ret']
spy_cummulative_returns.columns = columns

# Verify function worked by reviewing last few lines of DataFrame
spy_cummulative_returns.tail()
```
